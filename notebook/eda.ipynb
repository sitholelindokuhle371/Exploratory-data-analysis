```python
#importing all modules that will be used
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
#import csv file
df=pd.read_csv('data.csv')
df
```

# 2. Exploratory data analysis

```python
#clean the data by dropping all missing values
dfd=df.dropna()
df.dropna()
```

```python
#boxplot for Region and LifeExpectancy
lp=dfd[['Region','LifeExpectancy']]
sns.boxplot(lp)
plt.ylabel('Region')
plt.title('Life expectancy across different regions')
```

```python
#selecting Africa in region column
df1=dfd[dfd['Region']=='Africa']

#scatterplot for LiteracyRate and LifeExpectancy in Africa
plt.scatter(df1['LiteracyRate'],df1['LifeExpectancy'])
plt.xlabel('LiteracyRate (%)')
plt.ylabel('LifeExpectancy (%)')
plt.title('Relationship between Literacy and Life Expactancy in Africa')
plt.show()
```
